```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

%matplotlib inline

# Define the filenames and company names
file_names = [
    'AAPL_Historical_Data.csv',
    'GOOG_Historical_Data.csv',
    'MSFT_Historical_Data.csv',
    'AMZN_Historical_Data.csv'
]
company_name = ["APPLE", "GOOGLE", "MICROSOFT", "AMAZON"]

# Load data for each company and add to company_list
company_list = []
for file, name in zip(file_names, company_name):
    # Load stock data from CSV
    company_data = pd.read_csv(file)
    
    # Ensure the date column is named 'Date'
    if 'Date' in company_data.columns:
        company_data['Date'] = pd.to_datetime(company_data['Date'])
    else:
        print(f"Date column not found in {name} data!")
        continue  # Skip this company if the Date column is missing

    # Add a column to identify the company
    company_data["company_name"] = name
    
    # Filter the data for the specified date range
    filtered_data = company_data[(company_data['Date'] >= '2014-10-29') & (company_data['Date'] <= '2024-10-28')]
    
    # Append the filtered data to company_list
    company_list.append(filtered_data)

# Plotting the total volume of stock traded each day
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list):
    # Set the 'Date' as the index for each company's DataFrame
    company.set_index('Date', inplace=True)
    
    plt.subplot(2, 2, i + 1)  # Start subplots from 1
    company['Volume'].plot(title=f"Volume of Stocks Traded for {company_name[i]}")
    plt.ylabel('Volume')
    plt.xlabel('Date')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
ma_days = [10, 20, 50]

# Calculate moving averages for each company
for ma in ma_days:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()

# Create a plot for moving averages
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

# Plotting for each company using the DataFrame from company_list
for i, company in enumerate(company_list):
    axes[i // 2, i % 2].set_title(company['company_name'].iloc[0])  # Set title using the company name
    company[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[i // 2, i % 2])
    axes[i // 2, i % 2].grid(True) 

# Adjust layout
fig.tight_layout()
plt.show()
```

```python
# Calculate daily returns for each company
for company in company_list:
    company['Daily Return'] = company['Adj Close'].pct_change()

# Create a plot for daily returns
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

# Plotting for each company using the DataFrame from company_list
for i, company in enumerate(company_list):
    axes[i // 2, i % 2].plot(company['Daily Return'], linestyle='--', marker='o',markersize= 3, linewidth=1.5, label='Daily Return')
    axes[i // 2, i % 2].set_title(company['company_name'].iloc[0])
    axes[i // 2, i % 2].legend()
    axes[i // 2, i % 2].grid(True)  # Add grid to the subplot

# Adjust layout
fig.tight_layout()
plt.show() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the figure and call PairGrid on the DataFrame
return_fig = sns.PairGrid(tech_rets.dropna())

# Using map_upper to specify the upper triangle with scatter plots and light outline
return_fig.map_upper(plt.scatter, color='darkcyan', edgecolor='lightgray')

# Define the lower triangle with a kernel density estimate plot (kde) and light outline
return_fig.map_lower(sns.kdeplot, cmap='cool', fill=True, alpha=0.6)

# Define the diagonal as a series of histogram plots with light outline
return_fig.map_diag(plt.hist, color='purple',bins=30, edgecolor='lightgray')

# Add a dashed grid with reduced opacity to all axes
for ax in return_fig.axes.flatten():
    ax.grid(True, linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume closing_df is your DataFrame with closing prices already defined

# Set up the PairGrid for the DataFrame
returns_fig = sns.PairGrid(closing_df)

# Upper triangle with scatter plots (adjust marker size with 's')
returns_fig.map_upper(plt.scatter, color='purple', s=3)

# Lower triangle with KDE plots, using a color map and fill
returns_fig.map_lower(sns.kdeplot, cmap='cool') 

# Diagonal with histogram plots
returns_fig.map_diag(plt.hist, bins=30, edgecolor='lightgray')

# Add grid to all plots for better visibility
for ax in returns_fig.axes.flatten():
    ax.grid(True, linestyle='--', alpha=0.5)

# Set font size for x and y labels
for ax in returns_fig.axes.flatten():
    ax.set_xlabel(ax.get_xlabel(), fontsize=10)
    ax.set_ylabel(ax.get_ylabel(), fontsize=10)

# Add a title to the entire grid
returns_fig.fig.suptitle('Pairwise Relationships of Closing Prices for Tech Stocks', fontsize=13, y=1.02)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming tech_rets is already defined
rets = tech_rets.dropna()

area = np.pi * 20  # Adjust marker area

plt.figure(figsize=(10, 8))
plt.scatter(rets.mean(), rets.std(), s=area, alpha=0.6, color='purple')  # Add transparency and color
plt.xlabel('Expected Return', fontsize=14)
plt.ylabel('Risk (Standard Deviation)', fontsize=14)
plt.title('Risk vs Expected Return', fontsize=16)

# Annotate points
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x, y), xytext=(5, 5), textcoords='offset points', 
                 ha='right', va='bottom', fontsize=10, color='black', 
                 arrowprops=dict(arrowstyle='->', color='blue', lw=1))

plt.grid(True, linestyle='--', alpha=0.7)  # Add grid
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
